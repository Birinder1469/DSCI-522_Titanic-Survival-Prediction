```python
# Importing the required packages 

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree 
import graphviz
```

## Loading the dataset 

```python
# Loaded the training dataset 
titanic_data=pd.read_csv("..\\data\\train.csv")
```

## Training the model on the train.csv dataset 

```python
titanic_data.columns.values
```

```python
feature_cols=['Age','Fare','Pclass']
```

```python
titanic_model=DecisionTreeClassifier()
```

```python
titanic_data_dropped=titanic_data.dropna(subset=['Age','Fare','Pclass','Survived'])
```

```python
X=titanic_data_dropped.loc[:,feature_cols]
```

```python
y=titanic_data_dropped['Survived'].dropna()
```

```python
len(X)
```

```python
len(y)
```

```python
titanic_model.fit(X,y)
```

```python
titanic_model.predict(X)
```

```python
titanic_model.score(X,y)
```

## Plotting the Decision Tree for the trained dataset 

```python
feature_cols_3c=X.columns.values;

def save_and_show_decision_tree_3c(model, 
                                class_names = ['Killed', 'Survived'], 
                                save_file_prefix = '..\\results\\Titanic_decisionTree', **kwargs):
    """
    Saves the decision tree model as a pdf and a 
    """
    dot_data = tree.export_graphviz(model, out_file=None, 
                             feature_names=feature_cols_3c,  
                             class_names=class_names,  
                             filled=True, rounded=True,  
                             special_characters=True, **kwargs)  

    graph = graphviz.Source(dot_data) 
    graph.render(save_file_prefix) 
    return graph
graph = save_and_show_decision_tree_3c(titanic_model)
graph
```
